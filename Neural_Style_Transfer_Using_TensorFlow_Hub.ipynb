 # Neural Style Transfer Using TensorFlow Hub

ðŸ“Œ Problem Statement

How can we apply the artistic style of one image (e.g., a painting or texture) to another image (e.g., a photo of a temple), using deep learning?
Neural style transfer is a computer vision technique that uses deep learning to recombine the content of one image with the style of another. This allows us to create stunning, AI-generated artwork by blending photography and artistic styles.

ðŸŽ¯ Project Aim

To implement Neural Style Transfer using a pre-trained TensorFlow Hub model.
To take a content image and a style image, and output a new image with the content of the former and style of the latter.
To understand how deep convolutional neural networks can extract and recombine content/style features.


```python
# Step 1: Install and Import Dependencies
!pip install -q tensorflow tensorflow-hub matplotlib

import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Step 2: Load and Preprocess Images
def load_image(path_to_img, target_size=(512, 512)):
    img = Image.open(path_to_img).convert('RGB')
    img = img.resize(target_size)
    img = np.array(img) / 255.0
    img = img[np.newaxis, ...]
    return tf.convert_to_tensor(img, dtype=tf.float32)

content_image = load_image('content.png')
style_image = load_image('style.png')

# Step 3: Load Pre-trained Style Transfer Model from TensorFlow Hub
hub_model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')

# Step 4: Apply Style Transfer
stylized_image = hub_model(tf.constant(content_image), tf.constant(style_image))[0]

# Step 5: Display the Result
def imshow(image, title=None):
    image = image[0]  # remove batch dimension
    plt.imshow(image)
    if title:
        plt.title(title)
    plt.axis('off')

plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
imshow(content_image, "Content Image")

plt.subplot(1, 3, 2)
imshow(style_image, "Style Image")

plt.subplot(1, 3, 3)
imshow(stylized_image, "Stylized Image")

plt.tight_layout()
plt.show()
```

```python
# Step 2: Import necessary libraries
import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# Step 3: Load and resize content/style images

def load_image(path_to_img, target_size=(512, 512)):
    img = Image.open(path_to_img).convert('RGB')
    img = img.resize(target_size)
    img = np.array(img) / 255.0  # Normalize to [0,1]
    img = img[np.newaxis, ...]   # Add batch dimension
    return tf.convert_to_tensor(img, dtype=tf.float32)

# Load your images (make sure 'content.png' and 'style.png' are uploaded)
content_image = load_image('content.png')
style_image = load_image('style.png')
```

```python
# Step 4: Load the neural style transfer model from TensorFlow Hub
hub_model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
# Step 5: Apply style transfer to blend content and style
stylized_image = hub_model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
# Step 6: Display original content, style, and stylized image

def imshow(image, title=None):
    image = image[0]  # Remove batch dimension
    plt.imshow(image)
    if title:
        plt.title(title)
    plt.axis('off')

plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
imshow(content_image, "Content Image")

plt.subplot(1, 3, 2)
imshow(style_image, "Style Image")

plt.subplot(1, 3, 3)
imshow(stylized_image, "Stylized Image")

plt.tight_layout()
plt.show()
```
